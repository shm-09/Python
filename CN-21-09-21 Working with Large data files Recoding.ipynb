```python
def sfcoded(series):
  if math.isnan(series):
      return "missing"
  elif series >=50000:
      return "1"
  else:
      return "0"
```

```python
df_for = []
for i in range (df1['sitting_fees'].shape[0]):
    if math.isnan(df1['sitting_fees'][i]):
        df_for.append('missing')
    elif df1['sitting_fees'][i] >= 50000:
        df_for.append('1')
    else:
        df_for.append('0')
```

```python
df1.insert(2,"for_looped",df_for,True)
```
